Based on the graphs, color intensity seems to have the strongest relationship with alcohol

```python
A = np.array([[i,1] for i in data_wine.data[:,9]])  
theta = np.linalg.inv(A.T.dot(A)).dot(A.T.dot(y))
predx = np.linspace(min(data_wine.data[:,9]),max(data_wine.data[:,9]),200)
A_pred = np.array([[i,1] for i in predx])
pred_a = A_pred.dot(theta)

# Plotting
fig = figure(figsize=(20,10))
plot(predx,pred_a, label = 'Y= a$x + b')
scatter(data_wine.data[:,9],y,label = 'Alcohol')
xlabel('color_intensity',size = 20)
ylabel('Alcohol',size = 20)
xticks(size = 15)
yticks(size = 15)
legend(prop = {'size':20})
show()
```

```python
A_tri = np.array([[np.sin(i),1] for i in data_wine.data[:,9]])
theta_tri = np.linalg.inv(A_tri.T.dot(A_tri)).dot(A_tri.T.dot(y))
A_tri_pred = np.array([[np.sin(i),1] for i in predx])
pred_A_tri = A_tri_pred.dot(theta_tri)
# Plotting
fig = figure(figsize=(20,10))
plot(predx,pred_A_tri, label = 'Y= a$sin$x + b')
scatter(data_wine.data[:,9],y,label = 'Alcohol')
xlabel('color_intensity',size = 20)
ylabel('Alcohol',size = 20)
xticks(size = 15)
yticks(size = 15)
legend(prop = {'size':20})
show()
```

```python


# 2nd degree 
A2 = np.array([[i**2,i,1] for i in data_wine.data[:,9]])  
theta2 = np.linalg.inv(A2.T.dot(A2)).dot(A2.T.dot(y))
predx = np.linspace(min(data_wine.data[:,9]),max(data_wine.data[:,9]),200)
A2_pred = np.array([[i**2,i,1] for i in predx])
pred_a2 = A2_pred.dot(theta2)


# 3rd degree 
A3 = np.array([[i**3,i**2,i,1] for i in data_wine.data[:,9]])
theta3 = np.linalg.inv(A3.T.dot(A3)).dot(A3.T.dot(y))
A3_pred = np.array([[i**3, i**2,i,1] for i in predx])
pred_a3 = A3_pred.dot(theta3)


fig = figure(figsize=(15,10))
plot(predx,pred_a2, label = 'Y = a$x^2$ + b$x$ + c')
plot(predx,pred_a3, label = 'Y = a$x^3$ + b$x^2$ + c$x$ + d')
scatter(data_wine.data[:,9],y,label = 'Original Alcohol data')
xlabel('color_intensity',size = 20)
ylabel('Alcohol',size = 20)
xticks(size = 15)
yticks(size = 15)
legend(prop = {'size':20})
show()
```

```python
def rsquared(y,f):
    ## y is an array of observed values
    ## f is an array of predicted values
    ybar = np.mean(y)
    f = f.reshape(y.shape)
    ss_res = np.sum([i**2 for i in (y-f)])
    ss_reg = np.sum([i**2 for i in (f-ybar)])
    ss_tot = np.sum([i**2 for i in (y-ybar)])
    
    r2 = 1 - ss_res/ss_tot
    return r2
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y,
                                                   train_size = 0.75,
                                                   test_size = 0.25,
                                                   random_state = 1,
                                                   stratify = y)
## Fitting the model
classifier = KNeighborsClassifier(n_neighbors=3).fit(train_X,train_y)

## Predicting on training data
pred_train_y = classifier.predict(train_X)

## Predicting on test data
pred_test_y = classifier.predict(test_X)

## Printing the Accuracy on training and testing data
print('Accuracy on training data: ',sum(pred_train_y==train_y)/len(pred_train_y))
print('Accuracy on testing data: ',sum(pred_test_y==test_y)/len(pred_test_y))
```
