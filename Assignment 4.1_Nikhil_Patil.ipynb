Epidemiologists claim that the probability of breast cancer among Caucasian
women in their mid-50s is 0.005. 
An established test identified people who had
breast cancer and those that were healthy.
A new mammography test in clinical
trials has a probability of 0.85 for detecting cancer correctly. 
In women without
breast cancer, it has a chance of 0.925 for a negative result.

1. If a 55-year-old Caucasian woman tests positive for breast cancer, what is
the probability that she, in fact, has breast cancer?

2. Simulate this problem for estimating the probability in part 1 and provide
a plot of probability estimate (on Y-axis) vs total population size n (on
X-axis). Use n = 1000, 5000, 10000, 50000, 100000, 500000, 1000000.
1

```python
import numpy as np
from matplotlib.pyplot import *
```

# BAYES THEOREM

### $P(Have\ breast\ cancer\ |\ +\ test\ result) = \frac{P(+\ test\ result\ |\ Have\ breast\ cancer)\ \cdot\ P(Have\ breast\ cancer)}{P(+\ test\ result)}$

### P(+ test result) = P(+ test result | cancer)P(cancer) + P(+ test result | healthy)P(healthy)

```python
1-0.925 #False Positive
```

```python
1- 0.005 #Healthy 55 year olds
```

```python
dr= 0.85*0.005 + 0.07499999999999996*0.995
nr=0.85*0.005
prob= nr/dr
prob
```

### Generating a sample from this population

```python
n = [1000,5000, 10000, 50000, 100000,500000,1000000]
cancer = np.random.binomial(n,p=0.005)
healthy = n - cancer
```

### Testing

```python
## Number of women who have the breast cancer and tested positive
d_test_pos = np.random.binomial(cancer, p = 0.85)
## Number of women who are healthy and tested positive
h_test_pos = np.random.binomial(healthy, p = 0.07499999999999996)
```

```python
p2 = (d_test_pos + h_test_pos)/n
p1 = d_test_pos/n
```

```python
prob = p1/p2
prob
```

```python
style.use('dark_background')
fig = figure(figsize=(15,10))
ax = fig.add_subplot(1,1,1)
##Plotting Probability of samples being positive vs Sample population size n
ax.plot(n,prob)
ax.set_xlabel('Sample population size',size=15,)
ax.set_ylabel('Probability of positive samples',size=15)
ax.set_title('Probability of positive test vs Different population sizes', size=20)
```
